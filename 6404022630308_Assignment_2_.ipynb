1) Using NumPy, generate a 3x5 matrix filled with normally distributed random values with
mean 0 and standard deviation 1. Then, do the following:
1. Print the matrix


```python
import numpy as np

# สร้างเมทริกซ์
matrix = np.random.normal(0, 1, (3, 5))
print("1. เมทริกซ์ที่สร้างขึ้น:")
print(matrix)
```

2. Calculate the sum of all values in the matrix


```python
# คำนวณผลรวม
sum_of_values = np.sum(matrix)
print("\n2. ผลรวมของค่าทั้งหมดในเมทริกซ์:", sum_of_values)
```

3. Get the averages of the values in each column


```python
# คำนวณค่าเฉลี่ยของแต่ละคอลัมน์
column_averages = np.mean(matrix, axis=0)
print("\n3. ค่าเฉลี่ยของแต่ละคอลัมน์:", column_averages)
```

4. Find the minimum and maximum values in the matrix


```python
# ค้นหาค่าต่ำสุดและสูงสุด
min_value = np.min(matrix)
max_value = np.max(matrix)
print("\n4. ค่าต่ำสุดในเมทริกซ์:", min_value)
print("   ค่าสูงสุดในเมทริกซ์:", max_value)
```

5. Calculate the dot product of the matrix with its transpose

```python
# คำนวณผลิตภัณฑ์จุด
dot_product = np.dot(matrix, matrix.T)
print("\n5. ผลิตภัณฑ์จุดของเมทริกซ์กับเมทริกซ์สลับหลัง:")
print(dot_product)
```
